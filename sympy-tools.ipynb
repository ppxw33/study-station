````python
# !pip install sympy
import sympy as sp
# 最常用的 8 个功能
# 1. 定义符号变量（必须先做）
x, y, z = sp.symbols('x y z')  # 定义数学符号 x,y,z
a, b = sp.symbols('a b')
````

如果不先定义就会出现如下报错

````python
# 2. 代数化简
expr = (x**2 - 1) / (x - 1)
print(sp.simplify(expr))  # 输出 x + 1
````

````python
# 3. 展开 & 因式分解
# 展开
expr = (x+1)**3
print(sp.expand(expr))  # x³ + 3x² + 3x + 1

# 因式分解
expr = x**3 - 1
print(sp.factor(expr))  # (x - 1)(x² + x + 1)
````

````python
# 4. 解方程
# 解方程 x² - 5x + 6 = 0
print(sp.solve(x**2 -5*x +6, x))  # [2, 3]

# 解方程组
eq1 = x + y - 5
eq2 = x - y - 1
print(sp.solve((eq1, eq2), (x, y)))  # {x:3, y:2}
````

````python
# 5. 求导（微积分）
expr = sp.sin(x) * sp.exp(x)
print(sp.diff(expr, x))  # 求一阶导
print(sp.diff(expr, x, 2))  # 求二阶导
````

````python
# 6. 求积分
expr = x**2
print(sp.integrate(expr, x))  # 不定积分
print(sp.integrate(expr, (x, 0, 1)))  # 定积分 0~1
````

````python
# 7. 求极限
expr = sp.sin(x)/x
print(sp.limit(expr, x, 0))  # x→0 时极限 = 1
````

````python
# 8. 输出漂亮公式 / LaTeX
expr = sp.integrate(sp.sin(x), x)
sp.pprint(expr)  #打印公式展示
print(sp.latex(expr))
````
